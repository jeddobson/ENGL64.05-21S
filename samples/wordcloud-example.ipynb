```python
from wordcloud import WordCloud
```

```python
# load labMT sentiment scores
import csv
labMT_sentiment = dict()
with open('shared/engl64.05/lexicons/sentiment/word_sentiment_scores.csv', encoding = 'utf-8') as f:
    reader = csv.reader(f, delimiter = ',')
    for row in reader:
        labMT_sentiment[row[0]] = float(row[1])
```

```python
# select top twenty five positive words 
positive_words = sorted(labMT_sentiment, key=labMT_sentiment.__getitem__,reverse=True)[:25]
positive_words = { w : labMT_sentiment[w] for w in positive_words } 
```

```python
positive_wordcloud = WordCloud().generate_from_frequencies(positive_words)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(35,20),dpi=75)
plt.imshow(positive_wordcloud,interpolation="bilinear")
plt.axis("off")
plt.show()
```
